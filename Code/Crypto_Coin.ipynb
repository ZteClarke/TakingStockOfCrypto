```python
import pandas as pd 
from pandas_datareader import data as web 
import datetime as dt 
from pandasql import sqldf
sql = lambda q: sqldf(q, globals())
```

```python
start = dt.datetime(2018,1,1) 
end = dt.datetime(2022,12,7)

dogecoin = web.DataReader("DOGE-USD", "yahoo", start, end)
bitcoin = web.DataReader("BTC-USD", "yahoo", start, end)
ethereum = web.DataReader("ETH-USD", "yahoo", start, end)
spy = web.DataReader("SPY", "yahoo", start, end)
SP500 =  web.DataReader("^GSPC", "yahoo", start, end) 
```

```python
#for visualization 

import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
from matplotlib import style
import mplfinance as mpf
```

```python
spy["Daily Return"] = spy['Adj Close'].pct_change()
```

```python
# calculating the daily return
dogecoin["Daily Return"] = dogecoin['Adj Close'].pct_change()
```

```python
sql("SELECT * FROM dogecoin where Close = (select max(Close) from dogecoin)")
```

```python
plt.figure(figsize=(20, 20))
dogecoin['Adj Close'].plot()
plt.title("DOGE-USD Share Price")
plt.ylabel("Adj Close") 
plt.show()
style.use("ggplot") 
```

```python
plt.figure(figsize=(20, 20))
dogecoin['Daily Return'].plot()
plt.title("DOGE-USD Share Price")
plt.ylabel("Adj Close") 
plt.show()
style.use("ggplot") 
```
